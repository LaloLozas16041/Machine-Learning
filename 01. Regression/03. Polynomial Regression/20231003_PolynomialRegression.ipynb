### 03.- Regresión polinomial

### a) Importación de las librerías

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

### b) Importación del conjunto de datos

```python
dataset = pd.read_csv('Position_Salaries.csv')
X = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

### c) Entrenamiento del modelo de regresión lineal en todo el conjunto de datos

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X, y)
```

### d) Entrenamiento del modelo de regresión polinomial en todo el conjunto de datos

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 4)
X_poly = poly_reg.fit_transform(X)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly, y)
```

### d) Visualización de los resultados de la regresión lineal

```python
plt.scatter(X, y, color = 'red')
plt.plot(X, lin_reg.predict(X), color = 'blue')
plt.title('Regresión Lineal')
plt.xlabel('Nivel de la posición')
plt.ylabel('Salario')
plt.show()
```

### e) Visualización de los resultados de la regresión polinomial

```python
plt.scatter(X, y, color = 'red')
plt.plot(X, lin_reg_2.predict(poly_reg.fit_transform(X)), color = 'blue')
plt.title('Regresión Polinomial')
plt.xlabel('Nivel de la posición')
plt.ylabel('Salario')
plt.show()
```

### f) Visualización de los resultados de la regresión polinomial (para una resolución más alta y una curva más suave)

```python
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape((len(X_grid), 1))
plt.scatter(X, y, color = 'red')
plt.plot(X_grid, lin_reg_2.predict(poly_reg.fit_transform(X_grid)), color = 'blue')
plt.title('Regresión Polinomial')
plt.xlabel('Nivel de la posición')
plt.ylabel('Salario')
plt.show()
```

### g) Predecir un nuevo resultado con regresión lineal

```python
lin_reg.predict([[6.5]])
```

### h) Predecir un nuevo resultado con regresión polinomial

```python
lin_reg_2.predict(poly_reg.fit_transform([[6.5]]))
```

La regresión polinomial es una forma de regresión lineal en la que la relación entre la variable independiente $x$ y la variable dependiente $y$ se modela como un polinomio de grado $n$. A diferencia de la regresión lineal múltiple, que modela relaciones lineales con múltiples variables independientes, la regresión polinomial modela relaciones no lineales con una sola variable independiente. Aquí se explica paso a paso:

### 1. Modelo de Regresión Polinomial
El modelo se expresa como:
$y = \beta_0 + \beta_1x + \beta_2x^2 + \ldots + \beta_nx^n + \varepsilon$
donde:
- $y$ es la variable dependiente.
- $x$ es la variable independiente.
- $\beta_0, \beta_1, \ldots, \beta_n$ son los coeficientes del modelo que representan la intercepción y el efecto de $x$ elevado a diferentes potencias.
- $\varepsilon$ es el término de error.

### 2. Estimación de Coeficientes
Al igual que en la regresión lineal, los coeficientes se estiman utilizando el método de mínimos cuadrados ordinarios (MCO), minimizando la suma de los cuadrados de los errores entre los valores observados y los predichos por el modelo.

### 3. Transformación de Variables
Aunque el modelo es no lineal en $x$, es lineal en los términos de $x^2, x^3, \ldots, x^n$, lo que permite utilizar técnicas de regresión lineal. Esto se hace transformando $x$ en nuevas variables $x^2, x^3, \ldots, x^n$, y luego ajustando un modelo lineal múltiple.

### 4. Evaluación del Modelo
La evaluación del ajuste del modelo puede incluir:
- $R^2$ ajustado: una versión modificada de $R^2$ que penaliza la inclusión de términos adicionales innecesarios.
- Validación cruzada: para evaluar la capacidad predictiva del modelo y evitar el sobreajuste.
- Análisis de residuos: para verificar la adecuación del modelo y la presencia de patrones no lineales.

### 5. Selección del Grado del Polinomio
Una parte crucial es determinar el grado $n$ del polinomio:
- Un polinomio de grado demasiado bajo puede no capturar toda la curvatura de los datos (infraajuste).
- Un polinomio de alto grado puede ajustarse demasiado a los datos de entrenamiento, capturando el ruido en lugar de la relación subyacente (sobreajuste).
- Se utilizan métodos como la validación cruzada y el criterio de información de Akaike (AIC) para seleccionar el grado.

### 6. Supuestos de la Regresión Polinomial
Los supuestos son similares a los de la regresión lineal múltiple, pero con atención adicional al riesgo de sobreajuste y a la interpretación de los coeficientes, ya que estos ya no representan el cambio en $y$ por una unidad de cambio en $x$, sino en potencias de $x$.


### 7. Aplicaciones
La regresión polinomial se utiliza en situaciones donde la relación entre las variables no es simplemente lineal, como en el caso de la tasa de crecimiento de los tejidos, las trayectorias de vuelo, o en la economía cuando se estudian los retornos de escala.

### Ejemplo Práctico:
Supongamos que queremos modelar la relación entre la velocidad de un automóvil y el consumo de combustible, donde la relación no es lineal. Podríamos recolectar datos de velocidad y consumo de combustible, y ajustar un modelo polinomial para entender cómo el consumo de combustible aumenta con la velocidad a diferentes tasas, dependiendo de la potencia de la velocidad considerada en el modelo.

La regresión polinomial es muy útil para modelar relaciones complejas, y aunque es más complicada que la regresión lineal, sigue siendo una herramienta esencial en el análisis de regresión.
