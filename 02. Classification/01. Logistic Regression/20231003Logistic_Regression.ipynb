## 01.- Regresión logística

### a) Importación de las librerías

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

### b) Importación del conjunto de datos


```python
dataset = pd.read_csv('SNA.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

### c) Dividir el conjunto de datos en el conjunto de entrenamiento y el conjunto de prueba

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
print(X_train)
```

```python
print(y_train)
```

```python
print(X_test)
```

```python
print(y_test)
```

### d) Escalado de características

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
print(X_train)
```

```python
print(X_test)
```

### e) Entrenamiento del modelo de regresión logística en el conjunto de entrenamiento

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0)
classifier.fit(X_train, y_train)
```

### f) Predecir un nuevo resultado

```python
print(classifier.predict(sc.transform([[30,87000]])))
```

### g) Predicción de los resultados del conjunto de pruebas

```python
y_pred = classifier.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

### h) Hacer la matriz de confusión

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```

### i) Visualización de los resultados del conjunto de entrenamiento

```python
from matplotlib.colors import ListedColormap
X_set, y_set = sc.inverse_transform(X_train), y_train
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 10, stop = X_set[:, 0].max() + 10, step = 0.25),
                     np.arange(start = X_set[:, 1].min() - 1000, stop = X_set[:, 1].max() + 1000, step = 0.25))
plt.contourf(X1, X2, classifier.predict(sc.transform(np.array([X1.ravel(), X2.ravel()]).T)).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('salmon', 'dodgerblue')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1], c = ListedColormap(('salmon', 'dodgerblue'))(i), label = j)
plt.title('Regresión Logistica (Training set)')
plt.xlabel('Edad')
plt.ylabel('Salario Estimado')
plt.legend()
plt.show()
```

### j) Visualización de los resultados sobre el conjunto de prueba

```python
from matplotlib.colors import ListedColormap
X_set, y_set = sc.inverse_transform(X_test), y_test
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 10, stop = X_set[:, 0].max() + 10, step = 0.25),
                     np.arange(start = X_set[:, 1].min() - 1000, stop = X_set[:, 1].max() + 1000, step = 0.25))
plt.contourf(X1, X2, classifier.predict(sc.transform(np.array([X1.ravel(), X2.ravel()]).T)).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('salmon', 'dodgerblue')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1], c = ListedColormap(('salmon', 'dodgerblue'))(i), label = j)
plt.title('Regresión Logistica (Test set)')
plt.xlabel('Edad')
plt.ylabel('Salario Estimado')
plt.legend()
plt.show()
```

La Regresión Logística es un modelo estadístico utilizado para la clasificación y predicción de variables categóricas binarias (dos clases) o multinomiales (más de dos clases). A pesar de su nombre, no es un modelo de regresión en el sentido tradicional, sino un método de clasificación. A continuación, se proporciona una descripción detallada:

### 1. Modelo de Regresión Logística
La Regresión Logística utiliza la función logística (también conocida como función sigmoide) para modelar la probabilidad de que una observación pertenezca a una de las clases. La función logística se define como:

$$P(Y=1|X) = \frac{1}{1 + e^{-(\beta_0 + \beta_1X_1 + \ldots + \beta_pX_p)}}$$

donde:
- $P(Y=1|X)$ es la probabilidad de que la variable dependiente $Y$ sea igual a 1 dadas las variables independientes $X$.
- $\beta_0, \beta_1, \ldots, \beta_p$ son los coeficientes del modelo.
- $X_1, X_2, \ldots, X_p$ son las variables independientes.

### 2. Interpretación de los Coeficientes
Los coeficientes $\beta$ en la Regresión Logística representan el cambio logarítmico en la probabilidad de pertenecer a la clase 1 para un cambio unitario en la variable independiente correspondiente. Un coeficiente positivo aumenta la probabilidad de pertenecer a la clase 1, mientras que un coeficiente negativo la disminuye.

### 3. Función de Decisión
Para tomar una decisión de clasificación, se establece un umbral de probabilidad (por ejemplo, 0.5) y se compara con la probabilidad predicha por el modelo. Si la probabilidad predicha es mayor que el umbral, se clasifica como clase 1; de lo contrario, se clasifica como clase 0.

### 4. Entrenamiento del Modelo
El modelo de Regresión Logística se entrena utilizando el método de máxima verosimilitud, que encuentra los valores de los coeficientes $\beta$ que maximizan la probabilidad conjunta de observar los datos de entrenamiento bajo el modelo.

### 5. Regularización
Para evitar el sobreajuste, se pueden aplicar técnicas de regularización, como la regresión logística con regularización L1 (Lasso) o L2 (Ridge), que penalizan ciertos coeficientes, lo que puede llevar a la selección de características y un modelo más robusto.

### 6. Evaluación del Modelo
La evaluación de un modelo de Regresión Logística se realiza mediante métricas como la precisión, el área bajo la curva ROC (AUC-ROC), la sensibilidad, la especificidad y la matriz de confusión, dependiendo del problema de clasificación y los datos.

### 7. Casos de Uso
La Regresión Logística se utiliza en una amplia variedad de aplicaciones, como:
- Predicción de si un correo electrónico es spam o no spam.
- Predicción de si un paciente tiene una enfermedad (sí/no) basándose en características médicas.
- Clasificación de consumidores en categorías de riesgo crediticio (bueno/malo).
- Predicción de la probabilidad de éxito o fracaso de un proyecto.

### 8. Ventajas y Desventajas
- **Ventajas:**
  - Fácil de entender e interpretar.
  - No requiere muchas suposiciones sobre la distribución de los datos.
  - Puede manejar problemas de clasificación binaria y multinomial.
- **Desventajas:**
  - Puede no funcionar bien cuando la relación entre las variables independientes y la probabilidad es altamente no lineal.
  - Sensible a variables irrelevantes o multicolinealidad.

### Ejemplo Práctico:
Supongamos que deseas predecir si un estudiante será admitido o no en una universidad en función de sus calificaciones y su puntaje en un examen. Utilizando la Regresión Logística, puedes construir un modelo que tome las calificaciones y el puntaje del examen como variables independientes y prediga la probabilidad de admisión. Luego, puedes establecer un umbral para la probabilidad predicha para decidir si admitir o no al estudiante.

La Regresión Logística es un método fundamental en estadísticas y aprendizaje automático, y se utiliza ampliamente en la mayoría de las aplicaciones de clasificación binaria y multinomial.
