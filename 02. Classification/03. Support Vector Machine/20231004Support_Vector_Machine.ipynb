## 03.- Máquinas de soporte vectorial 

### a) Importación de las librerías

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

### b) Importación del conjunto de datos

```python
dataset = pd.read_csv('SNA.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

### c) Dividir el conjunto de datos en el conjunto de entrenamiento y el conjunto de prueba

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
print(X_train)
```

```python
print(y_train)
```

```python
print(X_test)
```

```python
print(y_test)
```

### d) Escalado de características

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
print(X_train)
```

```python
print(X_test)
```

### e) Entrenamiento del modelo SVM en el conjunto de entrenamiento

```python
from sklearn.svm import SVC
classifier = SVC(kernel = 'linear', random_state = 0)
classifier.fit(X_train, y_train)
```

### f) Predecir un nuevo resultado

```python
print(classifier.predict(sc.transform([[30,87000]])))
```

### g) Predicción de los resultados del conjunto de pruebas

```python
y_pred = classifier.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

### h) Hacer la matriz de confusión

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```

### i) Visualización de los resultados del conjunto de entrenamiento

```python
from matplotlib.colors import ListedColormap
X_set, y_set = sc.inverse_transform(X_train), y_train
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 10, stop = X_set[:, 0].max() + 10, step = 0.25),
                     np.arange(start = X_set[:, 1].min() - 1000, stop = X_set[:, 1].max() + 1000, step = 0.25))
plt.contourf(X1, X2, classifier.predict(sc.transform(np.array([X1.ravel(), X2.ravel()]).T)).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('salmon', 'dodgerblue')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1], c = ListedColormap(('salmon', 'dodgerblue'))(i), label = j)
plt.title('SVM (Training set)')
plt.xlabel('Edad')
plt.ylabel('Salario Estimado')
plt.legend()
plt.show()
```

### j) Visualización de los resultados del conjunto de pruebas

```python
from matplotlib.colors import ListedColormap
X_set, y_set = sc.inverse_transform(X_test), y_test
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 10, stop = X_set[:, 0].max() + 10, step = 0.25),
                     np.arange(start = X_set[:, 1].min() - 1000, stop = X_set[:, 1].max() + 1000, step = 0.25))
plt.contourf(X1, X2, classifier.predict(sc.transform(np.array([X1.ravel(), X2.ravel()]).T)).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('salmon', 'dodgerblue')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1], c = ListedColormap(('salmon', 'dodgerblue'))(i), label = j)
plt.title('SVM (Test set)')
plt.xlabel('Edad')
plt.ylabel('Salario Estimado')
plt.legend()
plt.show()
```

Las Máquinas de Soporte Vectorial (Support Vector Machines o SVM) son un conjunto de algoritmos de aprendizaje supervisado utilizados para la clasificación y regresión. Las SVM se destacan por su capacidad para encontrar límites de decisión óptimos en conjuntos de datos complejos y son ampliamente utilizadas en aplicaciones de aprendizaje automático. A continuación, se proporciona una descripción detallada de las Máquinas de Soporte Vectorial (SVM):

### 1. Fundamento del Algoritmo
El objetivo principal de una SVM es encontrar un hiperplano en un espacio de características (donde cada dimensión corresponde a una característica) que mejor separe las diferentes clases de datos. El hiperplano se selecciona de tal manera que maximiza el margen entre las clases, es decir, la distancia entre el hiperplano y los puntos de datos más cercanos de cada clase.

### 2. Hiperplano y Vectores de Soporte
- **Hiperplano:** El hiperplano es una superficie de decisión que separa las clases. En un problema de clasificación binaria, es un plano, pero en problemas de clasificación multiclase, puede ser un hiperplano multidimensional.
- **Vectores de Soporte:** Son los puntos de datos más cercanos al hiperplano. Estos puntos son cruciales para definir el hiperplano de separación y, por lo tanto, se denominan "vectores de soporte".

### 3. Función de Margen
La función de margen es fundamental en SVM. El margen se refiere a la distancia entre el hiperplano y los vectores de soporte. El objetivo de SVM es maximizar este margen. El hiperplano óptimo es aquel que maximiza la distancia entre clases y minimiza el error de clasificación.

### 4. Clasificación Lineal y No Lineal
Las SVM pueden realizar clasificaciones lineales y no lineales. En el caso de datos no linealmente separables, SVM utiliza una técnica llamada "kernel trick" para mapear los datos a un espacio dimensional superior donde son linealmente separables. Los tipos comunes de kernel incluyen el kernel lineal, el kernel polinomial y el kernel radial (RBF).

### 5. Parámetro de Regularización (C)
SVM introduce un parámetro de regularización $C$ que controla el equilibrio entre la maximización del margen y la minimización del error de clasificación. Un valor pequeño de $C$ da como resultado un margen más amplio pero permite un mayor error de clasificación, mientras que un valor grande de $C$ produce un margen más estrecho pero reduce el error de clasificación.


### 6. SVM en Problemas de Clasificación Multiclase
Para problemas de clasificación multiclase, SVM puede utilizarse en enfoques uno contra todos (One-Versus-All, OvA) o uno contra uno (One-Versus-One, OvO). En OvA, se entrena un clasificador binario para cada clase, mientras que en OvO, se entrena un clasificador binario para cada par de clases.



### 7. Ventajas y Desventajas
- **Ventajas:**
  - Efectivo en conjuntos de datos de alta dimensión.
  - Puede manejar problemas de clasificación lineales y no lineales.
  - Robusto frente a la presencia de datos atípicos (outliers).
  - La selección de kernel permite adaptarse a diversas formas de datos.
- **Desventajas:**
  - Requiere una elección adecuada del kernel y ajuste de hiperparámetros.
  - Puede ser computacionalmente costoso en conjuntos de datos grandes.
  - Interpretación menos intuitiva en comparación con otros algoritmos.



### 8. Ejemplo Práctico
Supongamos que deseas clasificar correos electrónicos como "spam" o "no spam" en función de sus palabras clave. Utilizando SVM, puedes entrenar un modelo que encuentre un hiperplano óptimo para separar los correos electrónicos en dos clases. Luego, puedes utilizar este modelo para clasificar nuevos correos electrónicos como spam o no spam en función de sus características.

En resumen, las Máquinas de Soporte Vectorial (SVM) son un poderoso algoritmo de aprendizaje automático utilizado tanto en problemas de clasificación como de regresión. Su capacidad para encontrar hiperplanos óptimos y manejar datos lineales y no lineales lo convierte en una herramienta valiosa en una amplia variedad de aplicaciones.
