## 07.- Clasificación Random Forest

### a) Importación de las librerías

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

### b) Importación del conjunto de datos

```python
dataset = pd.read_csv('SNA.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

### c) Dividir el conjunto de datos en el conjunto de entrenamiento y el conjunto de prueba

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
print(X_train)
```

```python
print(y_train)
```

```python
print(X_test)
```

```python
print(y_test)
```

### d) Escalado de características

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
print(X_train)
```

```python
print(X_test)
```

### e) Entrenamiento del modelo de clasificación Random Forest en el conjunto de entrenamiento

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 0)
classifier.fit(X_train, y_train)
```

### f) Predecir un nuevo resultado

```python
print(classifier.predict(sc.transform([[30,87000]])))
```

### g) Predicción de los resultados de la prueba

```python
y_pred = classifier.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

### h) Elaboración de la matriz de confusión

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```

### i) Visualización de los resultados del conjunto de entrenamiento

```python
from matplotlib.colors import ListedColormap
X_set, y_set = sc.inverse_transform(X_train), y_train
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 10, stop = X_set[:, 0].max() + 10, step = 0.25),
                     np.arange(start = X_set[:, 1].min() - 1000, stop = X_set[:, 1].max() + 1000, step = 0.25))
plt.contourf(X1, X2, classifier.predict(sc.transform(np.array([X1.ravel(), X2.ravel()]).T)).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('salmon', 'dodgerblue')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1], c = ListedColormap(('salmon', 'dodgerblue'))(i), label = j)
plt.title('Random Forest (Training set)')
plt.xlabel('Edad')
plt.ylabel('Salario')
plt.legend()
plt.show()
```

### j) Visualización de los resultados de las pruebas

```python
from matplotlib.colors import ListedColormap
X_set, y_set = sc.inverse_transform(X_test), y_test
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 10, stop = X_set[:, 0].max() + 10, step = 0.25),
                     np.arange(start = X_set[:, 1].min() - 1000, stop = X_set[:, 1].max() + 1000, step = 0.25))
plt.contourf(X1, X2, classifier.predict(sc.transform(np.array([X1.ravel(), X2.ravel()]).T)).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('salmon', 'dodgerblue')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1], c = ListedColormap(('salmon', 'dodgerblue'))(i), label = j)
plt.title('Random Forest (Test set)')
plt.xlabel('Edad')
plt.ylabel('Salario')
plt.legend()
plt.show()
```

La Clasificación con Bosques Aleatorios (Random Forest Classification) es un algoritmo de aprendizaje supervisado que se basa en la creación y combinación de múltiples árboles de decisión para mejorar la precisión de la clasificación. Un Bosque Aleatorio es una técnica de conjunto que utiliza múltiples árboles de decisión entrenados en diferentes subconjuntos de datos y características. A continuación, se proporciona una descripción detallada de la Clasificación con Bosques Aleatorios con un enfoque matemático:

### 1. Concepto de Bosque Aleatorio
Un Bosque Aleatorio es una colección de árboles de decisión individuales, cada uno entrenado en una porción diferente del conjunto de datos. Cada árbol se entrena en una selección aleatoria (con reemplazo) de ejemplos del conjunto de datos y un subconjunto aleatorio de características. La idea central detrás de un Bosque Aleatorio es que al combinar múltiples árboles de decisión independientes, se puede mejorar la precisión de la clasificación y reducir el sobreajuste.

### 2. Proceso de Construcción
El proceso de construcción de un Bosque Aleatorio implica los siguientes pasos:

#### a. Bootstrap Aggregating (Bagging)
Se genera un conjunto de datos de entrenamiento para cada árbol mediante el muestreo con reemplazo del conjunto de datos original. Esto crea múltiples conjuntos de datos llamados "conjuntos de entrenamiento bootstrap".

#### b. Selección de Características Aleatorias
En cada nodo de cada árbol, se realiza una selección aleatoria de un subconjunto de características para realizar la división. Esto ayuda a decorrelacionar los árboles y mejora la generalización.

#### c. Construcción de Árboles de Decisión
Se construyen múltiples árboles de decisión utilizando los conjuntos de entrenamiento bootstrap y las selecciones aleatorias de características. Cada árbol se construye mediante técnicas de árboles de decisión estándar, como el algoritmo CART.

#### d. Combinación de Predicciones
Para realizar una predicción en un nuevo dato, cada árbol en el Bosque Aleatorio emite una predicción. En la clasificación, se utiliza la "votación por mayoría" para combinar las predicciones de todos los árboles y decidir la clase final.

### 3. Ventajas de Random Forest
- **Reducción del Sobreajuste:** La combinación de múltiples árboles reduce la tendencia al sobreajuste que a menudo se observa en árboles de decisión individuales.
- **Mayor Precisión:** El promedio de las predicciones de múltiples árboles tiende a producir predicciones más precisas y estables.
- **Manejo de Características:** La selección aleatoria de características en cada árbol ayuda a manejar conjuntos de datos con muchas características y reduce la importancia de características irrelevantes o ruidosas.

### 4. Importancia de las Características
Random Forest proporciona una medida de la importancia de las características en la clasificación. Esta importancia se calcula observando cuánto cambia la precisión de la clasificación cuando las predicciones de una característica en particular se permutan aleatoriamente en el conjunto de datos de prueba. Las características que tienen un impacto significativo en la precisión tienen una mayor importancia.

### 5. Fórmula de Importancia de las Características
La importancia de una característica $X$ se puede calcular como sigue:

$$ \text{Importancia}(X) = \frac{1}{N} \sum_{t} \Delta \text{Accuracy}_t(X) $$

Donde:
- $N$ es el número de árboles en el Bosque Aleatorio.
- $\Delta \text{Accuracy}_t(X)$ es la diferencia en la precisión del árbol $t$ antes y después de la permutación aleatoria de las predicciones de la característica $X$.

### 6. Ejemplo Práctico
Supongamos que deseas clasificar correos electrónicos como "spam" o "no spam" en función de palabras clave en el texto del correo electrónico. Utilizando la Clasificación con Bosques Aleatorios, puedes construir un Bosque Aleatorio que consta de múltiples árboles de decisión, cada uno entrenado en un conjunto de datos de entrenamiento bootstrap y con selecciones aleatorias de características. Luego, puedes utilizar el Bosque Aleatorio para realizar clasificaciones en nuevos correos electrónicos.

En resumen, la Clasificación con Bosques Aleatorios es una técnica de aprendizaje automático que utiliza múltiples árboles de decisión independientes para mejorar la precisión de la clasificación. La combinación de predicciones de múltiples árboles y la selección aleatoria de características son aspectos clave de este algoritmo.
