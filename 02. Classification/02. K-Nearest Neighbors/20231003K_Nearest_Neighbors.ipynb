## 02.- K-Nearest Neighbors (K-NN)

### a) Importación de las librerías

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

### b) Importación del conjunto de datos

```python
dataset = pd.read_csv('SNA.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

### c) Dividir el conjunto de datos en el conjunto de entrenamiento y el conjunto de prueba

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
print(X_train)
```

```python
print(y_train)
```

```python
print(X_test)
```

```python
print(y_test)
```

### d) Escalado de características

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
print(X_train)
```

```python
print(X_test)
```

### e) Entrenamiento del modelo K-NN en el conjunto de entrenamiento

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2)
classifier.fit(X_train, y_train)
```

### f) Predecir un nuevo resultado

```python
print(classifier.predict(sc.transform([[30,87000]])))
```

### g) Predicción de los resultados del conjunto de pruebas

```python
y_pred = classifier.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

### h) Hacer la matriz de confusión

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```

### i) Visualizar los Resultados sobre el Conjunto de Entrenamiento

```python
from matplotlib.colors import ListedColormap
X_set, y_set = sc.inverse_transform(X_train), y_train
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 10, stop = X_set[:, 0].max() + 10, step = 1),
                     np.arange(start = X_set[:, 1].min() - 1000, stop = X_set[:, 1].max() + 1000, step = 1))
plt.contourf(X1, X2, classifier.predict(sc.transform(np.array([X1.ravel(), X2.ravel()]).T)).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('salmon', 'dodgerblue')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1], c = ListedColormap(('salmon', 'dodgerblue'))(i), label = j)
plt.title('K-NN (Conjunto de Entrenamiento)')
plt.xlabel('Edad')
plt.ylabel('Salario Estimado')
plt.legend()
plt.show()
```

### j) Visualizar los Resultados sobre el Conjunto de Prueba

```python
from matplotlib.colors import ListedColormap
X_set, y_set = sc.inverse_transform(X_test), y_test
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 10, stop = X_set[:, 0].max() + 10, step = 1),
                     np.arange(start = X_set[:, 1].min() - 1000, stop = X_set[:, 1].max() + 1000, step = 1))
plt.contourf(X1, X2, classifier.predict(sc.transform(np.array([X1.ravel(), X2.ravel()]).T)).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('salmon', 'dodgerblue')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1], c = ListedColormap(('salmon', 'dodgerblue'))(i), label = j)
plt.title('K-NN (Conjunto de Test)')
plt.xlabel('Edad')
plt.ylabel('Salario Estimado')
plt.legend()
plt.show()
```

El algoritmo de los k-vecinos más cercanos (k-Nearest Neighbors o k-NN) es un método de aprendizaje automático utilizado para la clasificación y la regresión. Su enfoque se basa en la suposición de que las observaciones similares tienden a estar en la misma categoría o tener valores de respuesta similares. A continuación, se proporciona una descripción detallada de k-NN:

### 1. Fundamento del Algoritmo
El principio fundamental de k-NN es que las observaciones se pueden clasificar según la mayoría de las clases a las que pertenecen sus vecinos más cercanos en el espacio de características. En otras palabras, si la mayoría de los k vecinos más cercanos a una observación de prueba pertenecen a una cierta clase, se predice que esa observación también pertenece a esa clase.

### 2. Parámetro k
El parámetro $k$ en k-NN representa el número de vecinos más cercanos que se considerarán al realizar una predicción. Este valor se elige antes de aplicar el algoritmo y tiene un impacto significativo en el rendimiento del modelo. Un valor pequeño de $k$ (por ejemplo, 1) puede llevar a un modelo muy ajustado a los datos de entrenamiento (sensible al ruido), mientras que un valor grande de $k$ puede suavizar las predicciones y reducir el efecto del ruido.

### 3. Clasificación con k-NN
Para la clasificación, k-NN calcula la distancia entre la observación de prueba y todas las observaciones de entrenamiento en el espacio de características. Las $k$ observaciones más cercanas (en términos de distancia) se seleccionan y se cuentan las clases a las que pertenecen. La clase más común entre los vecinos se asigna como la predicción para la observación de prueba.

### 4. Regresión con k-NN
Para la regresión, en lugar de contar las clases, se promedian los valores de respuesta de las $k$ observaciones más cercanas. La media se toma como la predicción para la observación de prueba.

### 5. Elección de la Función de Distancia
k-NN utiliza una función de distancia para medir la similitud entre observaciones. La distancia euclidiana es la métrica de distancia más comúnmente utilizada, pero también se pueden utilizar otras métricas, como la distancia de Manhattan o la distancia de Minkowski. La elección de la función de distancia depende del problema y los datos.

### 6. Normalización de Características
Es importante normalizar las características antes de aplicar k-NN, ya que las características con diferentes escalas pueden influir de manera desproporcionada en el cálculo de distancias. La normalización ayuda a asegurar que todas las características tengan el mismo peso en la clasificación.

### 7. Ventajas y Desventajas
- **Ventajas:**
  - Fácil de entender y aplicar.
  - No hace suposiciones sobre la distribución de los datos.
  - Puede ser utilizado para clasificación y regresión.
- **Desventajas:**
  - Sensible al ruido en los datos.
  - Puede ser computacionalmente costoso con grandes conjuntos de datos.
  - Requiere elegir el valor óptimo de $k$, lo que puede ser subjetivo.

### 8. Ejemplo Práctico
Supongamos que deseas clasificar correos electrónicos como "spam" o "no spam" basándote en características como la frecuencia de palabras clave. Utilizando k-NN, puedes calcular la distancia entre un nuevo correo electrónico y los correos electrónicos de entrenamiento previamente etiquetados. Luego, puedes seleccionar los $k$ correos electrónicos más cercanos y clasificar el nuevo correo electrónico según la mayoría de las etiquetas de esos correos electrónicos cercanos.

En resumen, k-NN es un algoritmo simple pero efectivo en el aprendizaje automático que se utiliza tanto para la clasificación como para la regresión. Su capacidad para hacer predicciones basadas en la similitud con observaciones cercanas lo hace adecuado para una variedad de problemas. Sin embargo, la elección del valor $k$ y la normalización de características son consideraciones importantes al aplicar k-NN.
