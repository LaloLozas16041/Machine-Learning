## 06.- Clasificación del árbol de decisión

### a) Importación de librerías

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

### b) Importación del conjunto de datos

```python
dataset = pd.read_csv('SNA.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

### c) División del conjunto de datos en conjunto de entrenamiento y conjunto de prueba

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
print(X_train)
```

```python
print(y_train)
```

```python
print(X_test)
```

```python
print(y_test)
```

### d) Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
print(X_train)
```

```python
print(X_test)
```

### e) Entrenamiento del modelo de clasificación de árbol de decisión en el conjunto de entrenamiento

```python
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(criterion = 'entropy', random_state = 0)
classifier.fit(X_train, y_train)
```

### f) Predecir un nuevo resultado

```python
print(classifier.predict(sc.transform([[30,87000]])))
```

### g) Predicción de los resultados de la prueba

```python
y_pred = classifier.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

### h) Elaboración de la matriz de confusión

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```

### i) Visualización de los resultados del conjunto de entrenamiento

```python
from matplotlib.colors import ListedColormap
X_set, y_set = sc.inverse_transform(X_train), y_train
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 10, stop = X_set[:, 0].max() + 10, step = 0.25),
                     np.arange(start = X_set[:, 1].min() - 1000, stop = X_set[:, 1].max() + 1000, step = 0.25))
plt.contourf(X1, X2, classifier.predict(sc.transform(np.array([X1.ravel(), X2.ravel()]).T)).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('salmon', 'dodgerblue')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1], c = ListedColormap(('salmon', 'dodgerblue'))(i), label = j)
plt.title('Árbol de Decisión (Training set)')
plt.xlabel('Edad')
plt.ylabel('Salario Estimado')
plt.legend()
plt.show()
```

### j) Visualización de los resultados del conjunto de prueba

```python
from matplotlib.colors import ListedColormap
X_set, y_set = sc.inverse_transform(X_test), y_test
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 10, stop = X_set[:, 0].max() + 10, step = 0.25),
                     np.arange(start = X_set[:, 1].min() - 1000, stop = X_set[:, 1].max() + 1000, step = 0.25))
plt.contourf(X1, X2, classifier.predict(sc.transform(np.array([X1.ravel(), X2.ravel()]).T)).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('salmon', 'dodgerblue')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1], c = ListedColormap(('salmon', 'dodgerblue'))(i), label = j)
plt.title('Árbol de Decisión  (Test set)')
plt.xlabel('Edad')
plt.ylabel('Salario Estimado')
plt.legend()
plt.show()
```

La Clasificación de Árbol de Decisión (Decision Tree Classification) es un algoritmo de aprendizaje supervisado utilizado para la clasificación de datos. Se basa en la creación de un árbol de decisiones que se utiliza para tomar decisiones de clasificación en función de las características de entrada. A continuación, se proporciona una descripción detallada de la Clasificación de Árbol de Decisión con un enfoque matemático:

### 1. Árbol de Decisión
Un árbol de decisión es una estructura jerárquica en forma de árbol que se utiliza para tomar decisiones. En el contexto de la clasificación, cada nodo interno del árbol representa una prueba en una característica, cada rama representa un resultado de la prueba, y cada hoja del árbol representa una clase o categoría de salida.

### 2. Criterio de Decisión
En cada nodo interno del árbol, se realiza una prueba en una característica específica para decidir qué rama seguir. El objetivo es dividir el conjunto de datos en subconjuntos más puros y homogéneos en términos de clases.

### 3. Entropía y Ganancia de Información
El concepto clave en la construcción del árbol de decisión es la entropía, que mide la impureza o la falta de homogeneidad en un conjunto de datos. Se utiliza la ganancia de información para medir cómo una prueba en una característica particular reduce la entropía del conjunto de datos. La ganancia de información se calcula mediante la fórmula:

$\text{Ganancia de Información} = \text{Entropía Inicial} - \text{Entropía Después de la Prueba}$

El objetivo es seleccionar la característica y el valor de corte que maximicen la ganancia de información.

### 4. Criterios de Selección de Características
En la práctica, se pueden utilizar varios criterios para seleccionar la característica y el valor de corte en cada nodo, incluyendo:
- Entropía (Criterio de Ganancia de Información).
- Gini impurity.
- Error de clasificación.

### 5. División Recursiva
El proceso de construcción del árbol se realiza de manera recursiva, dividiendo el conjunto de datos en subconjuntos más pequeños en cada nodo. El proceso continúa hasta que se cumple alguna condición de detención, como alcanzar una profundidad máxima, tener un número mínimo de muestras en un nodo o alcanzar una entropía mínima.

### 6. Podado del Árbol
Después de construir el árbol, es posible que se realice un proceso de podado para evitar el sobreajuste (overfitting). El podado implica eliminar nodos innecesarios para simplificar el árbol.

### 7. Clasificación
Una vez que se ha construido el árbol de decisión, se utiliza para realizar clasificaciones. Cada instancia de datos desciende por el árbol siguiendo las pruebas en cada nodo hasta llegar a una hoja, que representa la clase de salida.

### 8. Ventajas y Desventajas
- **Ventajas:**
  - Fácil de entender e interpretar.
  - Puede manejar datos categóricos y numéricos.
  - Proporciona información sobre la importancia de las características.
- **Desventajas:**
  - Puede ser propenso al sobreajuste si el árbol es demasiado profundo.
  - Sensible a pequeñas variaciones en los datos de entrenamiento.
  - Puede no funcionar bien en problemas con relaciones no lineales complejas.

### 9. Ejemplo Práctico
Supongamos que deseas clasificar correos electrónicos como "spam" o "no spam" en función de palabras clave en el texto del correo electrónico. Utilizando la Clasificación de Árbol de Decisión, puedes construir un árbol que realice pruebas en características como la frecuencia de palabras clave para tomar decisiones de clasificación.

En resumen, la Clasificación de Árbol de Decisión es un algoritmo de aprendizaje supervisado que utiliza árboles de decisión para realizar clasificaciones. La elección de características y criterios de división se basa en la ganancia de información y la entropía para maximizar la pureza de los subconjuntos de datos en cada nodo del árbol.
