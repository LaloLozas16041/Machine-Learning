## 02.- Agrupación jerárquica

### a) Importar las bibliotecas

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

### b) Importar el conjunto de datos

```python
dataset = pd.read_csv('Mall_Customers.csv')
X = dataset.iloc[:, [3, 4]].values
```

### c) Uso del dendrograma para encontrar el número óptimo de conglomerados

```python
import scipy.cluster.hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(X, method = 'ward'))
plt.title('Dendrograma')
plt.xlabel('Clientes')
plt.ylabel('Distancia Euclidiana')
plt.show()
```

### d) Entrenamiento del modelo de agrupación jerárquica en el conjunto de datos

```python
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters = 5, affinity = 'euclidean', linkage = 'ward')
y_hc = hc.fit_predict(X)
```

### e) Visualización de las agrupaciones

```python
plt.scatter(X[y_hc == 0, 0], X[y_hc == 0, 1], s = 100, c = 'red', label = 'Cluster 1')
plt.scatter(X[y_hc == 1, 0], X[y_hc == 1, 1], s = 100, c = 'blue', label = 'Cluster 2')
plt.scatter(X[y_hc == 2, 0], X[y_hc == 2, 1], s = 100, c = 'green', label = 'Cluster 3')
plt.scatter(X[y_hc == 3, 0], X[y_hc == 3, 1], s = 100, c = 'cyan', label = 'Cluster 4')
plt.scatter(X[y_hc == 4, 0], X[y_hc == 4, 1], s = 100, c = 'magenta', label = 'Cluster 5')
plt.title('Clusters de clientes')
plt.xlabel('Ingreso Anual (k$)')
plt.ylabel('Score de Gasto (1-100)')
plt.legend()
plt.show()
```

El Clustering Jerárquico (Hierarchical Clustering) es un enfoque de agrupamiento que busca construir una jerarquía de clústeres en un conjunto de datos. A diferencia de K-Means, que agrupa los datos en un número fijo de clústeres, el Clustering Jerárquico crea una estructura de árbol o dendrograma que muestra cómo los datos pueden agruparse en diferentes niveles de granularidad. A continuación, proporcionaré una descripción detallada del Clustering Jerárquico con enfoque matemático:

### 1. Conceptos Básicos
- $N$: El número de puntos de datos en el conjunto de datos.
- $X$: Conjunto de datos con $N$ puntos, donde $X = \{x_1, x_2, \ldots, x_N\}$.
- $D$: Matriz de distancias entre puntos de datos, donde $D_{ij}$ es la distancia entre $x_i$ y $x_j$.
- $Z$: Dendrograma que representa la jerarquía de clústeres.

### 2. Algoritmo de Clustering Jerárquico Aglomerativo
El algoritmo de Clustering Jerárquico Aglomerativo comienza con cada punto de datos como un clúster individual y procede de la siguiente manera:

#### a. Inicialización
Cada punto de datos $x_i$ se considera un clúster individual, y la matriz de distancias $D$ se inicializa con las distancias entre los puntos de datos.

#### b. Unión de Clústeres
En cada paso, se encuentran los dos clústeres más cercanos según la matriz de distancias $D$ y se combinan en un nuevo clúster. Esto se repite hasta que todos los puntos de datos estén en un solo clúster. La distancia entre los clústeres se puede calcular utilizando diferentes métodos, como enlace simple (single linkage), enlace completo (complete linkage), o enlace promedio (average linkage).

#### c. Construcción del Dendrograma
A medida que se unen los clústeres, se construye el dendrograma $Z$ que muestra la jerarquía de clústeres. Cada nodo interno del dendrograma representa la unión de dos clústeres, y las hojas del dendrograma representan los puntos de datos individuales.

### 3. Criterios de Enlace
Los criterios de enlace determinan cómo se calcula la distancia entre clústeres. Algunos de los criterios comunes son:

- **Enlace Simple (Single Linkage):** La distancia entre dos clústeres se define como la distancia más corta entre cualquier par de puntos, uno de cada clúster.

- **Enlace Completo (Complete Linkage):** La distancia entre dos clústeres se define como la distancia más larga entre cualquier par de puntos, uno de cada clúster.

- **Enlace Promedio (Average Linkage):** La distancia entre dos clústeres se define como el promedio de todas las distancias entre los puntos de los dos clústeres.

### 4. Dendrograma
El dendrograma es una representación visual de la jerarquía de clústeres. En el dendrograma, la altura de cada nodo interno representa la distancia entre los clústeres que se están fusionando. Los clústeres se pueden dividir en diferentes niveles de granularidad, lo que permite explorar la estructura jerárquica de los datos.

### 5. Elección del Número de Clústeres
Una ventaja del Clustering Jerárquico es que no es necesario especificar previamente el número de clústeres. El dendrograma permite seleccionar el número de clústeres observando en qué nivel se deben cortar las ramas del árbol para obtener el número deseado de clústeres.

### 6. Ventajas y Desventajas
- **Ventajas:**
  - No es necesario especificar el número de clústeres de antemano.
  - Proporciona una jerarquía de clústeres que puede ser útil para explorar la estructura de los datos.
- **Desventajas:**
  - Puede ser computacionalmente costoso en conjuntos de datos grandes.
  - La interpretación del dendrograma puede ser subjetiva.
  - Sensible a la elección de criterios de enlace.

### 7. Ejemplo Práctico
Supongamos que tienes datos de clientes de una tienda en línea y deseas segmentarlos en grupos basados en sus patrones de compra. Utilizando el Clustering Jerárquico, puedes construir un dendrograma que muestra cómo los clientes se agrupan en diferentes niveles de granularidad, desde grupos generales hasta grupos más específicos basados en su comportamiento de compra.

En resumen, el Clustering Jerárquico es una técnica de agrupamiento que crea una jerarquía de clústeres en un conjunto de datos. Es especialmente útil cuando no se conoce el número de clústeres de antemano y se desea explorar la estructura de los datos en diferentes niveles de detalle.
